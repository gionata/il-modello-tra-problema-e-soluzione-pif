```python
def algoritmo_greedy(oggetti_ordinati, cap):
    # Inizializza la soluzione all'insieme vuoto
    zaino = []
    peso_attuale = 0
    valore_attuale = 0
    
    print(f"{'Aggiunto':<12} | {'Peso Oggetto':<12} | {'Peso Totale':<12} | {'Valore Totale'}")
    print("-" * 60)
    
    # Cicla per ogni oggetto
    for obj in oggetti_ordinati:
        # Se il peso con inserendo l'oggetto è <= capacità
        if peso_attuale + obj['peso'] <= cap:
            zaino.append(obj)
            peso_attuale += obj['peso']
            valore_attuale += obj['valore']
            print(f"{obj['nome']:<7} | {obj['peso']:<12} | {peso_attuale:<12} | {valore_attuale}")
        else:
            # Altrimenti termina
            print("-" * 60)
            print(f"STOP: L'aggiunta di {obj['nome']} (peso {obj['peso']}) supererebbe la capacità!")
            break
            
    return zaino, peso_attuale, valore_attuale
```

```python
# solo per valurare il tempo di esecuzione

import time
import functools

# Definizione del decoratore 'timer'
def timer(func):
    """Decorator che stampa il tempo di esecuzione di una funzione."""
    @functools.wraps(func)
    def wrapper(*args, **kwargs):
        start_time = time.perf_counter()
        result = func(*args, **kwargs)
        end_time = time.perf_counter()
        
        execution_time = end_time - start_time
        print(f"La funzione {func.__name__!r} ha impiegato {execution_time:.6f} secondi.")
        return result
    return wrapper
```

```python
# Euristica greedy per PESO
@timer
def algoritmo_greedy_peso(oggetti, capacita_massima):
    return algoritmo_greedy(sorted(oggetti, key=lambda x: (x['peso'], -x['valore'])), capacita_massima)

soluzione, p_tot, v_tot = algoritmo_greedy_peso(oggetti, capacita_massima)

print(f"\nRISULTATO EURISTICA GREEDY:")
print(f"Numero oggetti scelti: {len(soluzione)}")
print(f"Peso finale: {p_tot} kg / {capacita_massima} kg")
print(f"Valore finale: {v_tot}")
```

```python
# Euristica greedy per VALORE
@timer
def algoritmo_greedy_valore(oggetti, capacita_massima):
    return algoritmo_greedy(sorted(oggetti, key=lambda x: (x['valore'], x['peso'])), capacita_massima)

soluzione, p_tot, v_tot = algoritmo_greedy_valore(oggetti, capacita_massima)
print(f"\nRISULTATO EURISTICA GREEDY:")
print(f"Numero oggetti scelti: {len(soluzione)}")
print(f"Peso finale: {p_tot} kg / {capacita_massima} kg")
print(f"Valore finale: {v_tot}")
```

```python
def upper_bound_dantzig(indice_partenza, peso_attuale, valore_attuale, capacita_residua, oggetti):
    """
    Calcola l'Upper Bound usando il rilassamento continuo.
    """
    # Creiamo una lista degli oggetti rimanenti
    rimanenti = oggetti[indice_partenza:]
    
    # Ordiniamo per densità di valore decrescente
    rimanenti_ordinati = sorted(rimanenti, key=lambda x: (-x['valore']/x['peso'], x['peso']))
    
    ub = valore_attuale
    spazio = capacita_residua
    
    for obj in rimanenti_ordinati:
        if obj['peso'] <= spazio:
            # Prendiamo l'oggetto intero
            spazio -= obj['peso']
            ub += obj['valore']
        else:
            # Prendiamo solo una frazione dell'oggetto critico
            frazione = spazio / obj['peso']
            ub += obj['valore'] * frazione
            break # Lo zaino è teoricamente pieno
            
    return ub

capacita_residua = 153
ub_teorico = upper_bound_dantzig(0, 0, 0, capacita_residua, oggetti)
print(f"Il valore massimo teorico (Upper Bound) è: {ub_teorico:.2f}")
```

```python
import math

def martello_toth_bound(indice_partenza, peso_corr, valore_corr, capacita_tot, oggetti_ordinati):
    """
    Calcola l'Upper Bound di Martello e Toth (UB2).
    Assume che oggetti_ordinati sia già ordinata per densità (valore/peso) decrescente.
    """
    n = len(oggetti_ordinati)
    peso_residuo = capacita_tot - peso_corr
    valore_limite = valore_corr
    
    # 1. Identificazione dell'oggetto critico 's'
    s = -1
    for i in range(indice_partenza, n):
        if oggetti_ordinati[i]['peso'] <= peso_residuo:
            peso_residuo -= oggetti_ordinati[i]['peso']
            valore_limite += oggetti_ordinati[i]['valore']
        else:
            s = i
            break
            
    # Se tutti gli oggetti rimanenti entrano nello zaino, il bound è il valore esatto
    if s == -1:
        return valore_limite

    # 2. Calcolo del bound di Dantzig (UB1) per confronto
    ub1 = valore_limite + math.floor(peso_residuo * (oggetti_ordinati[s]['valore'] / oggetti_ordinati[s]['peso']))

    # 3. Calcolo di UB2 (Martello e Toth)
    # Stima 1: Escludiamo l'oggetto critico s e proviamo a inserire s+1
    v_stima1 = valore_limite
    if s + 1 < n:
        v_stima1 += math.floor(peso_residuo * (oggetti_ordinati[s+1]['valore'] / oggetti_ordinati[s+1]['peso']))
    
    # Stima 2: Includiamo forzatamente l'oggetto s, togliendo valore da s-1
    v_stima2 = valore_limite + oggetti_ordinati[s]['valore'] - \
               math.ceil((oggetti_ordinati[s]['peso'] - peso_residuo) * (oggetti_ordinati[s-1]['valore'] / oggetti_ordinati[s-1]['peso']))

    # Il bound finale è il massimo tra le due stime (arrotondato per difetto perché i valori sono interi)
    ub2 = max(v_stima1, v_stima2)
    
    return ub2

oggetti_per_densita = sorted(oggetti, key=lambda x: (-x['valore']/x['peso'], x['peso']))

bound = martello_toth_bound(0, 0, 0, 153, oggetti_per_densita)
print(f"Upper Bound di Martello e Toth: {bound}")
```

```python
# Variabili per tracciare i progressi
miglior_valore = 0
soluzione_ottima = []
nodi_esplorati = 0

def branch_and_bound(indice, peso_att, valore_att, scelti, oggetti_ordinati) :
    global miglior_valore, soluzione_ottima, nodi_esplorati
    nodi_esplorati += 1

    # Aggiorna la soluzione incombente se ammissibile e migliore
    if valore_att > miglior_valore:
        miglior_valore = valore_att
        soluzione_ottima = scelti[:]

    # Caso base: fine oggetti
    if indice == len(oggetti_ordinati):
        return

    # Calcolo Upper Bound per il ramo corrente
    ub = martello_toth_bound(indice, peso_att, valore_att, capacita_massima, oggetti_ordinati)
    
    # POTATURA: Se il bound non può superare il valore della miglior soluzione già trovata, taglio il ramo
    if ub <= miglior_valore:
        return

    # BRANCHING (Strategia: Prima prova a includere, poi a escludere)
    # 1. Includi l'oggetto (se ci sta)
    if peso_att + oggetti_ordinati[indice]['peso'] <= capacita_massima:
        branch_and_bound(indice + 1, 
                         peso_att + oggetti_ordinati[indice]['peso'], 
                         valore_att + oggetti_ordinati[indice]['valore'], 
                         scelti + [oggetti_ordinati[indice]['nome']],
                         oggetti_ordinati)

    # 2. Escludi l'oggetto
    branch_and_bound(indice + 1, peso_att, valore_att, scelti, oggetti_ordinati)

@timer
def bb():
    # Preparazione: Ordinamento per densità
    oggetti_ordinati = sorted(oggetti, key=lambda x: (-x['valore']/x['peso'], x['peso']))
    
    # Esecuzione
    print("Avvio Branch and Bound...")
    branch_and_bound(0, 0, 0, [], oggetti_ordinati)

bb()
print(f"\n--- RISULTATI ---")
print(f"Valore Ottimo: {miglior_valore}")
print(f"Oggetti scelti: {soluzione_ottima}")
print(f"Nodi esplorati: {nodi_esplorati}")
```

```python
import math

@timer
def risolvi_zaino_mt1_iterativo(oggetti_input, capacita):
    # 1. Preparazione: Ordinamento per densità decrescente
    oggetti = sorted(oggetti_input, key=lambda x: (-x['valore']/x['peso'], x['peso']))
    n = len(oggetti)
    
    # Inizializzazione variabili di stato
    miglior_valore = 0
    soluzione_ottima = [0] * n
    
    corrente_valore = 0
    corrente_peso = 0
    scelte_attuali = [0] * n # 1 se incluso, 0 se escluso
    
    # Upper Bound di Martello e Toth semplificato per l'iterazione
    def calcola_ub2(idx, p_att, v_att):
        p_res = capacita - p_att
        v_lim = v_att
        s = -1
        for i in range(idx, n):
            if oggetti[i]['peso'] <= p_res:
                p_res -= oggetti[i]['peso']
                v_lim += oggetti[i]['valore']
            else:
                s = i
                break
        if s == -1: return v_lim
        # Logica Martello & Toth (UB2)
        v_s1 = v_lim + (p_res * oggetti[s+1]['valore'] // oggetti[s+1]['peso']) if s+1 < n else v_lim
        v_s2 = v_lim + oggetti[s]['valore'] - math.ceil((oggetti[s]['peso'] - p_res) * (oggetti[s-1]['valore']/oggetti[s-1]['peso']))
        return max(v_s1, v_s2)

    i = 0  # Puntatore all'oggetto corrente
    
    while True:
        # --- FASE FORWARD ---
        while i < n:
            if corrente_peso + oggetti[i]['peso'] <= capacita:
                # Se l'oggetto ci sta, lo includiamo
                corrente_peso += oggetti[i]['peso']
                corrente_valore += oggetti[i]['valore']
                scelte_attuali[i] = 1
            else:
                # Altrimenti lo saltiamo
                scelte_attuali[i] = 0
            i += 1
            
        # --- AGGIORNAMENTO SOLUZIONE INCOMBENTE ---
        if corrente_valore > miglior_valore:
            miglior_valore = corrente_valore
            soluzione_ottima = scelte_attuali[:]
            
        # --- FASE BACKTRACK ---
        # Cerchiamo l'ultimo oggetto che era stato incluso (scelte_attuali[j] == 1)
        i -= 1 # Torna all'ultimo indice valido
        while i >= 0 and scelte_attuali[i] == 0:
            i -= 1
            
        if i < 0:
            break # Abbiamo esplorato tutte le possibilità rilevanti
            
        # Rimuoviamo l'oggetto i-esimo dalla soluzione corrente per esplorare il ramo "senza i"
        corrente_peso -= oggetti[i]['peso']
        corrente_valore -= oggetti[i]['valore']
        scelte_attuali[i] = 0
        
        # --- PRUNING (POTATURA) ---
        # Verifichiamo se il ramo senza l'oggetto 'i' può portare a qualcosa di meglio
        if calcola_ub2(i + 1, corrente_peso, corrente_valore) <= miglior_valore:
            # Se l'Upper Bound è scarso, continuiamo a tornare indietro (backtrack ulteriore)
            # Questo simula il salto di interi rami dell'albero
            while i >= 0 and scelte_attuali[i] == 0:
                i -= 1
            if i < 0: break
            corrente_peso -= oggetti[i]['peso']
            corrente_valore -= oggetti[i]['valore']
            scelte_attuali[i] = 0
            
        i += 1 # Passiamo all'oggetto successivo dopo il backtrack

    return miglior_valore, soluzione_ottima

print("Avvio Branch and Bound...")
val_max, sol_ids = risolvi_zaino_mt1_iterativo(oggetti, capacita_massima)

print(f"\n--- RISULTATI ---")
print(f"Valore Ottimo: {miglior_valore}")
print(f"Oggetti scelti: {soluzione_ottima}")
print(f"Nodi esplorati: {nodi_esplorati}")
```
