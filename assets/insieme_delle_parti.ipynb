```python
oggetti = [
    {"nome": "Reflex", "peso": 5.2, "valore": 100},
    {"nome": "Barattolo", "peso": 2.3, "valore": 60},
    {"nome": "Pinne", "peso": 3.5, "valore": 70},
    {"nome": "Pallone", "peso": 1.5, "valore": 15}]

contatore_elementi = 0

def genera_insieme_parti_ricorsivo(indice, zaino_corrente):
    global contatore_elementi
    
    # CASO BASE: Abbiamo preso una decisione per tutti gli oggetti
    if indice == len(oggetti):
        contatore_elementi += 1
        
        # Calcolo peso e valore del sottoinsieme attuale
        nomi_scelti = [obj['nome'] for obj in zaino_corrente]        
        print(f"{nomi_scelti}")
        return

    # CASO RICORSIVO: Esploriamo i due rami dell'albero delle decisioni
    
    # 1. Decisione: NON includere l'oggetto corrente
    genera_insieme_parti_ricorsivo(indice + 1, zaino_corrente)
    
    # 2. Decisione: INCLUDERE l'oggetto corrente
    # Creiamo una copia della lista per non sporcare il ramo precedente
    genera_insieme_parti_ricorsivo(indice + 1, zaino_corrente + [oggetti[indice]])

# Avvio della ricorsione partendo dal primo oggetto (indice 0) e zaino vuoto []
genera_insieme_parti_ricorsivo(0, [])
```
